***Importing Libraries***

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
```

*Downloading Dataset*

```python
df = pd.read_csv(r"C:\Users\pc\Documents\pima-indians-diabetes.csv")
```

*Analyzing top 10 rows*

```python
df.head(10)
```

```python
df.columns=["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age","Outcome"]
df.head()
```

*Missing values*

```python
df.isnull().sum()
```

*There are no missing valuesin the dataset.*

```python
df['Insulin'].value_counts()
```

```python

sns.boxplot(x=df["Outcome"], y=df["Insulin"])
plt.title("Insulin Levels vs Diabetes Outcome")
plt.show()
```

***Data Visualization***

```python
df.hist(figsize= (12,8), bins = 30)
plt.suptitle("feature distribution", fontsize = 16)
plt.show()
```

***Boxpot for the detection of outliers***

```python
plt.figure(figsize=(12,10))
sns.boxplot(data=df)
plt.xticks
plt.title("Boxplots of Features to Detect Outliers")
plt.show()
```

***Correlation Matrix***

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

***Pairplot (How Features Relate to Each Other)***

```python
sns.pairplot(df, hue="Outcome", diag_kind="kde")
plt.show()
```

***Feature-Target Relationships (Violin & Boxplots)***

1. Glucose vs outcome

```python
sns.violinplot(x="Outcome", y="Glucose", data=df, inner="quartile")
plt.title("Glucose Levels vs Diabetes Outcome")
plt.show()
```

***BMI vs outcome***

```python
sns.violinplot(x="Outcome", y="BMI", data=df, inner="quartile")
plt.title("BMI vs Diabetes Outcome")
plt.show()
```

***Age vs outcome***

```python
sns.violinplot(x="Outcome", y="Age", data=df, inner="quartile")
plt.title("Age vs Diabetes Outcome")
plt.show()
```

***Countplot (Class Balance - Do We Have Enough Diabetic Cases?)***

```python
sns.countplot(x="Outcome", data=df)
plt.title("Class Distribution of Diabetes Outcome")
plt.show()
```
